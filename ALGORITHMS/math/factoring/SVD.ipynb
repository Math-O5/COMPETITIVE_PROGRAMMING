```
import numpy as np
from numpy import linalg as LA
```

```
# Projeção u em v
def projV(u, v):
  prVu = ((np.dot(u, v)/np.dot(v, v)) * v)
  return prVu
```

```
def gram_schmidt(a):
  n = len(a)
  q = []
  for i in range(n):
     v = a[i, :]
     for j in q:
       projection = projV(a[i, :], j)
       v =  v - projection
     q.append(v / LA.norm(v[:]))
  return q
```
